### Total Runtime

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Playlist size
data_size = [10000, 50000, 100000, 500000, 1000000]

# Baseline runtimes
baseline_runtimes = [1.6958, 25.5050, 57.1947, 290.1952, 592.7670]

# Scipy Sparse Matrix runtimes
scipy_sparse_runtimes = [0.0615, 0.3108, 0.5912, 6.4118, 13.0690]

# Scipy Sparse Matrix + Numba runtimes
scipy_sparse_numba_runtimes = [0.0490, 0.3115, 0.6238, 5.9389, 12.3004]

# lenskit Sparse Matrix + Numba runtimes
lenskit_sparse_numba_runtimes = [1.9775, 2.2486, 2.7586, 12.2411, 20.9515]

# Create the plot
plt.figure(figsize=(10, 6))
sns.set_style("darkgrid")
plt.plot(data_size, baseline_runtimes, label="Baseline")
plt.plot(data_size, scipy_sparse_runtimes, label="Scipy Sparse Matrix")
plt.plot(data_size, lenskit_sparse_numba_runtimes, label="Lenskit Sparse Matrix + Numba")
plt.plot(data_size, scipy_sparse_numba_runtimes, label="Scipy Sparse Matrix + Numba")
plt.title("Training Time")
plt.xlabel("Playlist Size")
plt.ylabel("Runtime (seconds)")
plt.xscale("log")
plt.yscale("log")
plt.xticks(data_size, labels=["10k", "50k", "100k", "500k", "1m"])
plt.legend()
plt.savefig('train.jpg')
```
